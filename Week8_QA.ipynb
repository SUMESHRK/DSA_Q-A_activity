# The manager wants to find out whether the same amount was spent for the three advertisements (TV , Radio and Newspaper)

```python
#check whether amount spend for all modes of advertisements are same or not using oneway Anova

#H0= Amount spend for three advertisements are same,µ1=µ2=µ3
#Ha= Amount spent for atleast one advertisement differs from the rest,different µ values

import scipy.stats as stats
fval,pval = stats.f_oneway(newdata['TV'],newdata['Radio'],newdata['Newspaper'])
#print(fvalue,pvalue)

alpha = 0.05  # set the significance level
if pval > alpha:
    print("Accept the null hypothesis")
    print("Same amount was spend for the three modes of advertisements.")
elif pval < alpha:
    print("Reject the null hypothesis")
    print("Different amount was spend for the three modes of advertisements.")
```
